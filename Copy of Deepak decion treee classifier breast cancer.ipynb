```python
from pandas import read_csv,get_dummies,DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
dataset =read_csv("/content/drive/MyDrive/Datasets/Breast_Cancer.csv")
dataset['A Stage']=dataset['A Stage'].map({'Regional':1,'Distant':0})
dataset['Estrogen Status']=dataset['Estrogen Status'].map({'Positive':1,'Negative':0})
dataset['Progesterone Status']=dataset['Progesterone Status'].map({'Positive':1,'Negative':0})
dataset['Status']=dataset['Status'].map({'Alive':1,'Dead':0})
data2=get_dummies(dataset,['Race','Marital Status','T Stage','N Stage','6th Stage','differentiate','Grade'])
x=data2.drop('Status',axis=1)
y=data2['Status']
x_=StandardScaler().fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x_,y,test_size=.20)
x_train,y_train=SMOTE(random_state=66).fit_resample(x_train,y_train)
```

```python
#Regular method ( method#1)
from sklearn import tree
Breast_can_class =tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = 7)
Breast_can_class.fit(x_train, y_train)
y_pred = Breast_can_class.predict(x_test)
```

```python
from sklearn import metrics
Accuracy=metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", Accuracy)
con_matrix = metrics.confusion_matrix(y_test, y_pred)
print (con_matrix)
recall = metrics.recall_score(y_test, y_pred)
print ("Recall :",recall)
percision=metrics.precision_score(y_test, y_pred)
print("Percesion :",percision)
```

```python
from sklearn.model_selection import GridSearchCV
Breast_can_class2 = tree.DecisionTreeClassifier(criterion = 'entropy')
depth = {'max_depth': [1,2,3,4,5,10,12,18,24,30,36]}
grid_search1 = GridSearchCV(estimator=Breast_can_class2, param_grid=depth, scoring='recall', cv=7)
grid_search1.fit(x_, y )
best_depth = grid_search1.best_params_
print(best_depth)
best_result = grid_search1.best_score_
print(best_result)
```
