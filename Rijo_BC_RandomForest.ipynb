```python
# Data Preparation
from pandas import read_csv, get_dummies,Series
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
Dataset=read_csv('/content/drive/MyDrive/DataSets/Breast_Cancer.csv')
Dataset['A Stage']=Dataset['A Stage'].map({'Regional':1,'Distant':0})
Dataset['Estrogen Status']=Dataset['Estrogen Status'].map({'Positive':1,'Negative':0})
Dataset['Progesterone Status']=Dataset['Progesterone Status'].map({'Positive':1,'Negative':0})
Dataset['Status']=Dataset['Status'].map({'Alive':1,'Dead':0})
Dataset2=get_dummies(Dataset,['Race','Marital Status','T Stage','N Stage','6th Stage','differentiate','Grade'])
x=Dataset2.drop('Status', axis=1)
y=Dataset2['Status']
x_scaled=StandardScaler().fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split( x_scaled, y, test_size = 0.20, random_state = 121)
x_train,y_train =SMOTE (random_state = 101).fit_resample(x_train,y_train)
```

```python
# Regular method ( method#1)
from sklearn import ensemble
Rijo_classifier1 =ensemble.RandomForestClassifier(n_estimators=100, criterion='entropy', max_features='auto', random_state=1)
Rijo_classifier1.fit(x_train,y_train)
y_pred1=Rijo_classifier1.predict(x_test)
```

```python
# Evaluation
# Acuracy and confusion matrix
from sklearn import metrics
Accuracy=metrics.accuracy_score(y_test, y_pred1)
print("Accuracy: ", Accuracy)
con_matrix = metrics.confusion_matrix(y_test, y_pred1)
print (con_matrix)
recall = metrics.recall_score(y_test, y_pred1)
print (recall)
percision=metrics.precision_score(y_test, y_pred1)
print(percision)
```

```python
# GridSearchCV (method#2)
from sklearn.model_selection import GridSearchCV
Rijo_classifier2 = ensemble.RandomForestClassifier(criterion='entropy', max_features='auto', random_state=1)
no_trees = {'n_estimators': [400, 450, 500, 550, 600, 650]}
grid_search1 = GridSearchCV(estimator=Rijo_classifier2, param_grid=no_trees, scoring='recall', cv=8)
grid_search1.fit(x_scaled, y)
best_parameters = grid_search1.best_params_
print(best_parameters)
best_result = grid_search1.best_score_
print(best_result)
```

```python
# Regular method with best parameter ( method#1)
Rijo_classifier3 = ensemble.RandomForestClassifier(n_estimators=450, criterion='entropy', max_features='auto', random_state=1)
Rijo_classifier3.fit(x_train,y_train)
y_pred3=Rijo_classifier3.predict(x_test)
imp_features = Series(Rijo_classifier3.feature_importances_, index=list(x)).sort_values(ascending=False)
print(imp_features)
```

```python
# Evaluation
# Acuracy and confusion matrix
from sklearn import metrics
recall=metrics.recall_score(y_test, y_pred3)
print("recall: ", recall)
con_matrix = metrics.confusion_matrix(y_test, y_pred3)
print (con_matrix)
```

```python
# Using pipeline (method #3)
from imblearn.pipeline import Pipeline
Rijo_classifier6 = Pipeline([('balancing', SMOTE(random_state = 101)),
        ('classification', ensemble.RandomForestClassifier(criterion='entropy', max_features='auto', random_state=1) )])
no_trees = {'classification__n_estimators': [20,40,60,80,100,120]}
grid_search3 = GridSearchCV(estimator=Rijo_classifier6, param_grid=no_trees, scoring='recall', cv=8)
grid_search3.fit(x_scaled, y)

best_parameters = grid_search3.best_params_
print(best_parameters)
best_result = grid_search3.best_score_
print(best_result)
```
